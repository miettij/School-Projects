```python
import pandas as pd
import numpy as np
b = pd.read_csv('temperature_min.csv', sep=";", header=None)

print(b)
```

```python
import pystan
import stan_utility
```

```python
temp = np.asarray(b)[:,:]
#print(temp)
x = temp[1:,0].astype(int)
y_HKI = temp[1:,1].astype(float)
y_JYV = temp[1:,2].astype(float)
y_HAI = temp[1:,3].astype(float)
y_ROV = temp[1:,4].astype(float)
y_ALL = temp[1:,1:].astype(float)
y_ALL = y_ALL.flatten()
M = len(y_ALL)
N = len(x)
xpred = 2050
y = [y_HKI, y_JYV, y_OUL, y_ROV]


data = dict(
        N=38,
        x=x,
        y_HKI = y_HKI,
        y_JYV = y_JYV,
        y_HAI = y_HAI,
        y_OUL = y_ROV,
        y_ALL = y_ALL,
        M = M,
        xpred = 2200)
print(data)
```

```python
model = stan_utility.compile_model('oldhier5.stan')
```

```python
fit = model.sampling(data=data, seed=194838)
```

```python
print(fit)
```

```python
samples = fit.extract(permuted=True)
# preview 500 posterior samples of (alpha, beta)
plt.scatter(samples['alpha_HKI'][:500], samples['beta_HKI'][:500], 5)
plt.xlabel('alpha_HKI')
plt.ylabel('beta_HKI');
```

```python
plt.scatter(samples['alpha_JYV'][:500], samples['beta_JYV'][:500], 5)
plt.xlabel('alpha_JYV')
plt.ylabel('beta_JYV');
```

```python
print('Pr(beta_HKI > 0) = {}'.format(np.mean(samples['beta_HKI'] > 0)))
print('Pr(beta_JYV > 0) = {}'.format(np.mean(samples['beta_JYV'] > 0)))
print('Pr(beta_HAI > 0) = {}'.format(np.mean(samples['beta_HAI'] > 0)))
print('Pr(beta_ROV > 0) = {}'.format(np.mean(samples['beta_OUL'] > 0)))
```
